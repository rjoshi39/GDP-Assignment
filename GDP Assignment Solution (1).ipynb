Instead of dropping them we can replace %growth over previous year with the mean of %growth over previous year and use that to calculate that gsdp of that year.

Taking mean of gsdp will be wrong since it will not capture the trend (increasing or decreasing of gdsp)

We will first draw best fit lines for all states and then clean the data, since it is asked to use all available data to draw the graphs

```python
#we will use following defined functions to find best fit line of the points

def mean(x):
    return sum(x)/len(x)



def best_fit_slope_and_intercept(xs,ys):
    xsbar=sum(xs)/len(xs)
    ysbar=sum(ys)/len(ys)
    xys=xs*ys
    xysbar=sum(xys)/len(xys)
    xxs=xs*xs
    xxsbar=sum(xxs)/len(xxs)
    m = (((xsbar*ysbar) - xysbar) /
         ((xsbar*xsbar) - xxsbar))
    
    b = ysbar - m*xsbar
    
    return m, b

#m is slope and b is intercept, you can read more about the formula online:
# link: https://stackoverflow.com/questions/22239691/code-for-best-fit-straight-line-of-a-scatter-plot-in-python
```

```python
#we also make a list of all states 
states=['Andhra Pradesh',
        'Arunachal Pradesh', 'Assam', 'Bihar', 'Chhattisgarh', 
        'Goa', 'Gujarat', 'Haryana', 'Himachal Pradesh', 'Jammu & Kashmir', 'Jharkhand', 
        'Karnataka', 'Kerala', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya',
        'Mizoram', 'Nagaland', 'Odisha', 'Punjab', 'Rajasthan', 'Sikkim', 
        'Tamil Nadu', 'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand', 'West Bengal', 'India']
states.sort()

#note that we also added India in the states

#we also rename last column in gdp_current_prices as India
c=list(gdp_current_price.columns)
c[-1]='India'
gdp_current_price.columns=c
```

```python
#we now iterate over states and plot a bestfit line for all
for column in gdp_current_price:
    # ignore the ('Items Description','Duration')
    if column in states:
        plotdata=gdp_current_price[gdp_current_price['Items  Description']=='(% Growth over previous year)'][column]
        #first slice ['Items  Description']=='(% Growth over previous year)'
        #and then take data of state from it
        xs=np.array(plotdata.index)
        ys=plotdata.values
        #take only those values which are not null
        xs = xs[~np.isnan(ys)]
        ys = ys[~np.isnan(ys)]
        
        #now use best fit function to find slope and intercept
        m, b = best_fit_slope_and_intercept(xs,ys)
        #find the points on line to plot the line
        regression_line = [(m*x)+b for x in xs]
        #draw scatter and regression line plot on same line
        f, ax=plt.subplots()
        plt.scatter(xs,ys,color='#003F72')
        plt.plot(xs, regression_line)
        #give lables to xticks of the years
        plt.xticks(np.arange(min(xs), min(xs)+len(xs), 1.0))
        labels = [item.get_text() for item in ax.get_xticklabels()]
        if(len(xs)>5):
            labels[5] = '2016-17'
        if(len(xs)>0):
            labels[0] = '2011-12'
        if(len(xs)>1):
            labels[1] = '2012-13'
        if(len(xs)>2):
            labels[2] = '2013-14'
        if(len(xs)>3):
            labels[3] = '2014-15'
        if(len(xs)>4):
            labels[4] = '2015-16'
        ax.set_xticklabels(labels)
        #make appropriate lables 
        plt.xlabel('Year')
        plt.ylabel('(% Growth over previous year)')
        plt.suptitle(column)
        plt.show()
```

Please Note that the above graphs will not be used to draw any insights, This is to illustrate that not all graphs or data will give you some new insights always, This is why its called "Explotratory" Data Analysis. However this insight of drawing best fit lines can now be used to plot gsdp current prices plots and slope of that plot can be used to compare growth rates of different states. Below is the same code as above except for gsdp current prices.

We will also save the growthrate(slope of best fit line) for each state in a dictionary and then use it to perform the said data cleaning operations described earlier, dropping row 5 and 10 and handling missing values of year 2015-16 by growthrates.

You might choose to handle missing values by taking averages, remember you will take average of only %Growthrate over previous year and use that to calculate the gsdp current prices of 2015-16, check further for more details.

```python
growthrate={}
for column in gdp_current_price:
    # ignore the ('Items Description','Duration')
    if column in states:
        plotdata=gdp_current_price[gdp_current_price['Items  Description']=='GSDP - CURRENT PRICES (` in Crore)'][column]
        xs=np.array(plotdata.index)
        ys=plotdata.values
        xs = xs[~np.isnan(ys)]
        ys = ys[~np.isnan(ys)]
        m, b = best_fit_slope_and_intercept(xs,ys)
        growthrate[column]=m
        regression_line = [(m*x)+b for x in xs]
        f, ax=plt.subplots()
        plt.scatter(xs,ys,color='#003F72')
        plt.plot(xs, regression_line)
        plt.xticks(np.arange(min(xs), min(xs)+len(xs), 1.0))
        labels = [item.get_text() for item in ax.get_xticklabels()]
        if(len(xs)>5):
            labels[5] = '2016-17'
        if(len(xs)>0):
            labels[0] = '2011-12'
        if(len(xs)>1):
            labels[1] = '2012-13'
        if(len(xs)>2):
            labels[2] = '2013-14'
        if(len(xs)>3):
            labels[3] = '2014-15'
        if(len(xs)>4):
            labels[4] = '2015-16'
        ax.set_xticklabels(labels)
        plt.xlabel('Year')
        plt.ylabel('GSDP - CURRENT PRICES (` in Crore)')
        plt.suptitle(column)
        plt.show()        
```

Now we use growthrate dictionary to finally handle the missing values of row 4 and 9 

and then drop row 5 and 10

here growthrate has average increase or growth in gsdp current prices per year, however this is absolute increase and not a normalised scale, for example state of sikkhim has gsdp of 100cr (for eg) and its growth rate is 30cr/year, where as another state Maharashtra has a gsdp of 1000cr and has a growthrate of 100cr/year. Clearly absolute growth is greater of Maharashtra, but sikkhim grew more on normalised scale i.e it grew 30% while maharashtra grew 10%

Hence we will also calculate "Average growth rate" which will be average of %Growth over previous year, this will be a normalized scale

We will show analysis using both here

We first clean the data by dropping using growthrate to handle missing value sin row 4 and 9. Then drop rows 5 and 10 and then make a row for average growth rate as average of all %growt over previous year. Finally drop all the UT from the data.

## Plot % contribution of primary, secondary and tertiary

```python
# creating a time slot for various time period of day
def gdp_percentage(gsdp, total_gsdp):
    return gsdp/total_gsdp
    
gsva['primary_gdp_percent'] = gsva[['Primary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
gsva['secondary_gdp_percent'] = gsva[['Secondary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
gsva['tertiary_gdp_percent'] = gsva[['Tertiary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
```

```python
#plt.rcParams["figure.figsize"] = [20,10]
gsva = gsva.sort_values('Per Capita GSDP (Rs.)')
gsva[['primary_gdp_percent', 'secondary_gdp_percent', 'tertiary_gdp_percent']].plot(kind='bar', stacked=True)
gsva[["Population ('00)"]].plot(kind='bar')
```

```python
values=per_capita.values
total_states=len(values)
gsva['percentile score']=gsva['Per Capita GSDP (Rs.)'].apply(lambda x: sum(values<x)*100/total_states)
#now finding corealtion
gsva['percentile score'].corr(gsva['primary_gdp_percent'])
```
